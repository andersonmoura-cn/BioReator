```python
import sympy as sp

# apresentação
from IPython.display import display, Math
from sympy import latex

# utils
from utils.cleanDisplay import clean
from utils.saveFT import save_mimo_tf

# valores
from config.config import values
```

```python
# variável de Laplace
s = sp.symbols('s')

# elementos de A
a1, a2, a3, a4 = sp.symbols('a1 a2 a3 a4')

# elementos de B
b1, b2, b3, b4 = sp.symbols('b1 b2 b3 b4')

# matrizes
A = sp.Matrix([[a1, a2],
               [a3, a4]])

B = sp.Matrix([[b1, b2],
               [b3, b4]])

I = sp.eye(2)

# expressão geral
G = (s*I - A).inv() * B

# variáveis físicas
Mi_m, S, D, Ks, Y, X, Sf = sp.symbols('Mi_m S D Ks Y X Sf')

# substituições
subs_dict = {

    a1: Mi_m*S/(Ks+S) - D,
    a2: X*Mi_m*Ks/(Ks+S)**2,

    a3: -Mi_m*S/(Y*(Ks+S)),
    a4: -D - (Mi_m*X/Y)*(Ks/(Ks+S)**2),

    b1: -X,
    b2: 0,
    b3: Sf - S,
    b4: D
}

G_subs = G.subs(subs_dict)

# ponto de operação
S_bar, X_bar = sp.symbols('S_bar X_bar')

G_eq = G_subs.subs({
    S: S_bar,
    X: X_bar
})

G_final = sp.simplify(G_eq)

# sp.pprint(G_final)
# print(latex(G_eq))
display(G_final)
```

```python
# puxando valores de config.py
values_sym = {
    sp.Symbol(k): v for k, v in values.items()
}

G_num = sp.simplify(G_eq.subs(values_sym))

# sp.pprint(G_num)
outputs = ["X", "S"]
inputs = ["D", "Sf"]

# limpando a saída
G_clean = G_num.applyfunc(clean)

for i, y in enumerate(outputs):
    for j, u in enumerate(inputs):
        display(Math(f"G_{{{y},{u}}}(s) = {latex(G_clean[i,j])}"))
```
