```python
import control as ctrl
import numpy as np
from config.matrizes_config import matrizes
```

```python
A = matrizes['A']
B = matrizes['B']
C = matrizes['C']
```

Análise de controlabilidade

```python
matriz_controlabilidade = ctrl.ctrb(A, B)
print("Matriz de Controlabilidade:\n", matriz_controlabilidade)
print("Posto da matriz de controlabilidade:", np.linalg.matrix_rank(matriz_controlabilidade))
print("Sistema controlável?", "Sim "if np.linalg.matrix_rank(matriz_controlabilidade) == A.shape[0] else "Não")
```

Análise de observabilidade

```python
matriz_observabilidade = ctrl.obsv(A, C)
print("\nMatriz de Observabilidade:\n", matriz_observabilidade)
print("Posto da matriz de observabilidade:", np.linalg.matrix_rank(matriz_observabilidade))
print("Sistema observável?", "Sim" if np.linalg.matrix_rank(matriz_observabilidade) == A.shape[0] else "Não")
```
